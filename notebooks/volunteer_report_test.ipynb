# Nova `get_result_per_class`

```python
vr.df_test_answers[m_study_phase_images].groupby("image_type").count()
```

```python
m_study_phase_images = (
    (vr.df_test_answers["phase_name"] == "test_phase")
    & (vr.df_test_answers["in_study_phase"] == "True")
    & (vr.df_test_answers["event_type"] == "answer")
) 

df_study_phase_images = vr.df_test_answers[m_study_phase_images]
df_study_phase_images.head()
```

```python
m_catch_image = (
    (vr.df_test_answers["phase_name"] == "test_phase")
    & (vr.df_test_answers["in_study_phase"] == "False")
    & (vr.df_test_answers["image_type"] == "catch_stimuli")
) 

df_catch_image = vr.df_test_answers[m_catch_image]
df_catch_image
```

```python
df_images = concat([df_study_phase_images, df_catch_image]).sort_values(by="event_seq")
df_images
```

```python
df_result_per_class = (
    df_images
        .groupby("image_class")
        .agg({
            "is_right_answer": "sum",
            "event_seq":"count"
        })
        .rename({"event_seq":"count"}, axis=1)
)

df_result_per_class["right_perc"] = (
    100.0*(df_result_per_class["is_right_answer"]
    / df_result_per_class["count"])
).round(1)

df_result_per_class
```
